```python
# На основе таблицы можно просмотреть данные и выделить какие-то гипотезы для будуших исследований. Например, что те, у кого родители живут отдельно, имеют больше прогулов в школе. 
```

```python
# проверим датасет на наличие аномальных значений.

df_hr['anomally_inc'] = np.where(
                            df_hr['absences'] > 
                            np.mean(df_hr.absences) + 3*np.std(df_hr.absences),                 
                            1, 0)
```

```python
df_anomally_inc = df_hr[df_hr['anomally_inc'] == 1]

df_anomally_inc
```

```python
#получилось, что шесть персоналий имеют аномальные значения в количестве прогулов. Проверим далее. 
```

```python
print("Count of Anomally absencese is {} ".format(len(df_anomally_inc)))
```
